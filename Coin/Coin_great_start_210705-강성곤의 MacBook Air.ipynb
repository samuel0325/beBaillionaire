```python
def up_down_market(ticker):
    past = pybithumb.get_ohlcv(ticker)
    m5 = past['close'].rolling(5).mean()
    current_p = pybithumb.get_current_price(ticker)
    last_mean5 = m5[-2]
    print(ticker,"current_price :",current_p)
    print("지난 5일",ticker,"평균 :",last_mean5)
    if current_p > last_mean5:
        return True
    else:
        return False
```

```python
up_down = up_down_market("ETH")
if up_down:
    print("떡상 가즈아!!!!!")
else:
    print("돔황챠!!!!!")
```

```python
import asyncio
```

```python
async def make_americano():
    print("아메리카노 시작")
    await asyncio.sleep(3)
    print("아메리카노 끝")
    
async def make_latte():
    print("라떼 시작")
    await asyncio.sleep(5)
    print("라떼 끝")
    
async def main():
    coco1 = make_americano()
    coco2 = make_latte()
    await asyncio.gather(coco1,coco2)
    
print("카페 영업 시작!")
asyncio.run(main())
print("카페 영업 종료!")
```
